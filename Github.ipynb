## Creating the github Repository



1. Create a private repository on GitHub

Go to GitHub

Click New repository.

Enter a repository name.

Select Private.

Click Create repository.


Do not initialize it with a README if your local folder already contains files.

2. Open your folder in VS Code

In VS Code, select:

File → Open Folder

Choose the folder you want to upload.

3. Open the terminal in VS Code

Press:

Ctrl + `

or:

Terminal → New Terminal

4. Initialize Git

Run:

`git init`

5. Add all files

`git add .`

6. Create the first commit

`git commit -m "Initial commit"`


If Git asks for your identity:

`git config --global user.name "Your Name"`

`git config --global user.email "your_email@example.com"`

Then run the commit command again.

7. Connect to the GitHub repository

Copy the repository URL from GitHub. It will look like:

https://github.com/username/my-private-repo.git

Then run:

git remote add origin https://github.com/username/my-private-repo.git

8. Push to GitHub

Rename the branch to main:

`git branch -M main`


Push:

`git push -u origin main`

GitHub may prompt you to authenticate in your browser.

Future updates

Whenever you change files:

`git add .`

`git commit -m "Describe changes"`

`git push`

Machine learning is a game of two parts:
1. Get data into a numerical representations
2. Build a model to learn patterns in that numerical representations 

To showcase this, let's create some 'known' data using linear regression formula. (straight line with known parameters)

```python
#create 'known' parameters
weight = 0.7 #weight, bias are just names for a and b in y =ax + b linear regression formula
bias = 0.3

#create
start = 0
end = 1
step = 0.02
X = torch.arange(start, end, step).unsqueeze(dim=1) #unsqueeze to add an extra dimension
y = weight * X + bias #X is capital for tensor and y is vector

X[:10], y[:10]
```

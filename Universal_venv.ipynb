Installing the Universal virtual Environment

1. Install miniforge3 
Miniforge3-26.1.1-3-Windows-x86_64

2. Create a reusable environment

Open Miniforge Prompt (or a terminal where Conda is available):

 > conda create -n universal python=3.12

Activate it:

 > conda activate universal

Verify:

 > python --version
3. Install packages

For a general-purpose environment:

 > conda install numpy pandas matplotlib scipy jupyter

4. See your environments
 > conda env list

Example output:

 > base
 > universal

5. Use the environment in VS Code
Open VS Code.
Install the Python extension if you haven't already.
Press Ctrl + Shift + P.
Select Python: Select Interpreter.
Choose the interpreter from your Miniforge environment, something like:
...\miniforge3\envs\universal\python.exe

**Grid Search**
It is a method used to systematically find the best hyperparameters for a machine learning model.

Machine learning models have hyperparameters (settings you choose before training), such as:

- n_estimators in a Random Forest

- C in Logistic Regression

- max_depth in Decision Trees

Choosing the right combination improves model performance.

Instead of manually trying combinations, GridSearch tests all possible combinations in a defined grid.

```python
mod = KNeighborsRegressor().fit(X, y)
pipe = Pipeline([('scale', StandardScaler()), ('model', KNeighborsRegressor(n_neighbors=1))])
#pipe.get_params()
```

```python
mod = GridSearchCV(estimator=pipe,
                   param_grid={'model__n_neighbors': [1, 2, 3, 4, 5]},
                   cv=3)
```

```python
mod.fit(X, y);
pd.DataFrame(mod.cv_results_)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
df = pd.read_csv("D:\dev\Learning-Python\drawndata1.csv")
df.head(3)
```

```python
X = df[['x', 'y']].values
y = df['z'] == "a"
```

```python
#To access the columns as arrays for plotting, convert X to a NumPy array (X.values)or use pandas'(X.iloc) accessor:
plt.scatter(X[:, 0], X[:, 1], c=y)
```

*Applying a Scalar*

```python
from sklearn.preprocessing import StandardScaler, QuantileTransformer

X_new = QuantileTransformer(n_quantiles=100).fit_transform(X)
plt.scatter(X_new[:, 0], X_new[:, 1], c=y)

#X and Y axis a numerically similar, so the distance between points is not dominated by one of the axes.
```

*Function to deal with the plotting*

```python
def plot_output(scaler):
    pipe = Pipeline([
        ('scale', scaler), 
        ('model', KNeighborsClassifier(n_neighbors=20, weights='distance'))
    ])

    pred = pipe.fit(X, y).predict(X)

    plt.figure(figsize=(9, 3))
    plt.subplot(131)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.title("Original Data")
    plt.subplot(132)
    X_tfm = scaler.transform(X)
    plt.scatter(X_tfm[:, 0], X_tfm[:, 1], c=y)
    plt.title("Transformed Data")
    plt.subplot(133)
    X_new = np.concatenate([
        np.random.uniform(0, X[:, 0].max(), (5000, 1)), 
        np.random.uniform(0, X[:, 1].max(), (5000, 1))
    ], axis=1)
    y_proba = pipe.predict_proba(X_new)
    plt.scatter(X_new[:, 0], X_new[:, 1], c=y_proba[:, 1], alpha=0.7)
    plt.title("Predicted Data")
```

```python
plot_output(scaler=StandardScaler())
```

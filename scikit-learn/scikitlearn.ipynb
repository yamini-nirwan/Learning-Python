**Grid Search**
It is a method used to systematically find the best hyperparameters for a machine learning model.

Machine learning models have hyperparameters (settings you choose before training), such as:

- n_estimators in a Random Forest

- C in Logistic Regression

- max_depth in Decision Trees

Choosing the right combination improves model performance.

Instead of manually trying combinations, GridSearch tests all possible combinations in a defined grid.

```python
mod = KNeighborsRegressor().fit(X, y)
pipe = Pipeline([('scale', StandardScaler()), ('model', KNeighborsRegressor(n_neighbors=1))])
#pipe.get_params()
```

```python
mod = GridSearchCV(estimator=pipe,
                   param_grid={'model__n_neighbors': [1, 2, 3, 4, 5]},
                   cv=3)
```

```python
mod.fit(X, y);
pd.DataFrame(mod.cv_results_)
```

*Function to deal with the plotting*

This function shows how different feature scalers affect a K-Nearest Neighbors (KNN) classifier by plotting three views: original data, scaled data, and predictions on new points.

StandardScaler → mean 0, variance 1
MinMaxScaler → values between 0–1

*Transformed Data*: This plot shows how scaling changes the geometry of the dataset.
This matters because KNN relies on distances.

*Predicted Data*: KNN predicts class probabilities.
Each random point is colored by the predicted probability.
Interpretation:

- Dark colors → strong class-1 probability
- Light colors → strong class-0 probability

This visualizes the decision boundary of the KNN classifier.

```python
def plot_output(scaler):
   

    pipe = Pipeline([
        ('scale', scaler), 
        ('model', KNeighborsClassifier(n_neighbors=20, weights='distance')) #weights='distance'→ closer neighbors count more strongly than farther ones.
    ])

    pred = pipe.fit(X, y).predict(X) 
    """
    # Train the Pipeline: 
    # fit(X, y): fits the scaler, scales the data, trains the KNN model
    # predict(X): predicts labels for the training data
   """
    plt.figure(figsize=(9, 3))  #3 side-by-side plots.
    plt.subplot(131)
    plt.scatter(X[:, 0], X[:, 1], c=y) #c=y → color points according to class labels
    plt.title("Original Data")

    plt.subplot(132)
    X_tfm = scaler.transform(X)
    plt.scatter(X_tfm[:, 0], X_tfm[:, 1], c=y)
    plt.title("Transformed Data")
    plt.subplot(133)
    
    X_new = np.concatenate([
        np.random.uniform(0, X[:, 0].max(), (5000, 1)), 
        np.random.uniform(0, X[:, 1].max(), (5000, 1))
    ], axis=1)
    y_proba = pipe.predict_proba(X_new)
    plt.scatter(X_new[:, 0], X_new[:, 1], c=y_proba[:, 1], alpha=0.7)
    plt.title("Predicted Data")
```

This code demonstrates why scaling matters for KNN.

KNN uses distance calculations: $d = \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}$
​
If one feature has larger scale:

- feature1: 0–1000
- feature2: 0–1

Then feature1 dominates the distance, making feature2 almost irrelevant.

Scaling fixes that.

```python
plot_output(scaler=StandardScaler())
```

```python
plot_output(scaler=QuantileTransformer(n_quantiles=100))
```
